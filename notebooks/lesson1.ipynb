## Load packages

We'll use packages to help us to different things. Packages are built on functions. Functions take an input, do something with it, and give you an output. Packages will save us time because someone else did all the thinking in how to clean up the data and give it to us in a useable form.

Some packages help us manipulate the data. Other packages help us make charts. For Spotify data, there's already a package called `spotipy` that helps us interact with developer data within Spotify. But, another package caled `tidyspotify` simplifies that even more!

```python
import altair as alt
import dotenv
import os
import pandas as pd
import spotipy
import tidyspotify

from spotipy.oauth2 import SpotifyClientCredentials
from tidyspotify import get_artist_audio_features

#import warnings
#warnings.simplefilter("ignore")

dotenv.load_dotenv()

config = {
    'client_id': os.environ["SPOTIPY_CLIENT_ID"],
    'client_secret': os.environ["SPOTIPY_CLIENT_SECRET"],
}
                                
sp = spotipy.Spotify(client_credentials_manager = SpotifyClientCredentials(**config))

#tidyspotify.save_credentials()
```

```python
sample_dfs = {}

select_artists = ["rihanna", "taylor swift", 
                  "hamilton", "dear evan hansen", 
                  "ed sheeran", "ariana grande"]

for i in select_artists:
    name = i.replace(" ", "_")
    sample_dfs[name] = tidyspotify.get_artist_audio_features(i)
    
final = pd.DataFrame()
for key, value in sample_dfs.items():
    final = final.append(value)
    
final = final.reset_index(drop=True)
final.to_csv("./data/artist_audio_features.csv", index=False)

unique_artists = final.artist_uri.drop_duplicates().to_list()

album_dfs = {}
for i in unique_artists:
    album_dfs[i] = tidyspotify.get_artist_albums(artist_id=i)
    
final2 = pd.DataFrame()
for key, value in album_dfs.items():
    final2 = final2.append(value)
    
final2.reset_index(drop=True).to_csv("./data/artist_albums.csv", index=False)
```

```python
def make_scatterplot(df, x_col, y_col):
    chart = (alt.Chart(df)
             .mark_circle()
             .encode(
                 x=f"{x_col}",
                 y=f"{y_col}"
             )
            )
    display(chart)
```
